```python
from tensorflow import keras
from keras.models import Sequential
from keras.layers import Conv2D
from keras.layers import MaxPooling2D
from keras.layers import Flatten
from keras.layers import Dense
import keras.utils as image
import pandas as pd
```

```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
from PIL import Image
import os
```

```python
classifier=Sequential()
classifier.add(Conv2D(32,(3,3), input_shape=(128,128,3),activation="relu"))
classifier.add(Conv2D(32,(3,3),activation="relu"))
classifier.add(MaxPooling2D(pool_size=(2,2)))
```

```python
classifier.add(Flatten())
classifier.add(Dense(units=128,activation="relu"))
classifier.add(Dense(units=6,activation="softmax"))
```

```python
classifier.compile(loss='categorical_crossentropy', optimizer="adam", metrics=['accuracy'])
```

```python
from keras.preprocessing.image import ImageDataGenerator
train_datagen=ImageDataGenerator(rescale=1/255)
validation_datagen=ImageDataGenerator(rescale=1/255)
```

```python
train_dataset=train_datagen.flow_from_directory("tinkers/train",target_size=(128,128),batch_size=64,class_mode='categorical')
validation_dataset=validation_datagen.flow_from_directory("tinkers/validation",target_size=(128,128),batch_size=64,class_mode='categorical')
test_dataset=validation_datagen.flow_from_directory("tinkers/test",target_size=(128,128),batch_size=128,class_mode='categorical')
```

```python
train_dataset.class_indices
```

```python
history=classifier.fit(train_dataset, steps_per_epoch=15,epochs=5,validation_data=validation_dataset)
```

```python
for i in range(1,6):
    dir_path = "tinkers/test/Label"+str(i)
    for j in os.listdir(dir_path):
            img = Image.open(os.path.join(dir_path, j)).convert('RGB').resize((128, 128))
            plt.imshow(img)
            plt.show()
            X = image.img_to_array(img)
            X = np.expand_dims(X, axis=0)
            val = np.argmax(classifier.predict(X))
            print(f"The image belongs to class:{i}")
            print(f"The model predicts it belongs to class:{val+1}")
            if val+1 == i:
                print("Correct prediction")
            else:
                print("Incorrect prediction")
```

```python
from sklearn.metrics import classification_report, confusion_matrix
test_loss, test_accuracy = classifier.evaluate(test_dataset)
print("Test Loss:", test_loss)
print("Test Accuracy:", test_accuracy)
```

```python
y_pred = classifier.predict(test_dataset)
y_pred = np.argmax(y_pred, axis=1)
y_true = test_dataset.labels
```

```python
# Generate classification report and confusion matrix
print("Classification Report:")
print(classification_report(y_true, y_pred))

```

```python
print("Confusion Matrix:")
print(confusion_matrix(y_true, y_pred))
```
